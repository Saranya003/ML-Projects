```python
#train and Test 
from sklearn.model_selection import train_test_split
# data # label
```

```python
x=df.iloc[:,df.columns != 'Outcome'] #data
y=df.iloc[:,df.columns == 'Outcome'] #label
```

```python
print(y)
```

```python
xtrain, xtest, ytrain, ytest= train_test_split(x, y, test_size=0.2)# 20% test
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
from sklearn.ensemble import RandomForestClassifier # import 
```

```python
model=RandomForestClassifier()
```

```python
model.fit(xtrain, ytrain)# fit is function to train the alg
```

```python
predict_output= model.predict(xtest)# to test the alg
predict_output
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(predict_output, ytest)
print("Accuracy_Score", acc)
```
